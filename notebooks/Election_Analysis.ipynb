```python
rename_map = {
    "प्रदेश": "province",
    "जिल्ला": "district",
    "प्रतिनिधि सभा निर्वाचन क्षेत्र": "constituency",
    "उम्मेदवारको नाम": "candidate",
    "लिङ्ग": "gender",
    "राजनीतिक दल / स्वतन्त्र": "party",
    "चुनाव चिन्ह": "symbol",
    "जम्मा प्राप्त मत": "votes",
    "स्थान": "rank",
    "कैफियत": "remarks",
}

df = df.rename(columns=rename_map)
```

```python
import re

new_columns = []

for col in df.columns:
    col = str(col)              # Convert to string
    col = col.strip()           # Remove leading/trailing spaces
    col = col.lower()           # Convert to lowercase
    col = re.sub(r"\s+", "_", col)  # Replace spaces with underscores
    new_columns.append(col)

df.columns = new_columns

print(list(df.columns))
```

```python
# strip whitespace on object columns
for c in df.select_dtypes(include="object").columns:
    df[c] = df[c].astype(str).str.strip()

# replace blank-ish strings with NaN
df = df.replace({"": np.nan, "nan": np.nan, "None": np.nan})

# drop rows missing the key info (adjust if needed)
required = [c for c in ["province", "district", "constituency", "candidate", "party"] if c in df.columns]
df = df.dropna(subset=required)

print("After dropping missing core fields:", df.shape)
```

```python
# Step 1: Clean and convert "votes" column to numeric
if "votes" in df.columns:

    # Convert to string (in case it's mixed types)
    df["votes"] = df["votes"].astype(str)

    # Remove commas (example: "12,345" → "12345")
    df["votes"] = df["votes"].str.replace(",", "", regex=False)

    # Remove spaces (example: "12 345" → "12345")
    df["votes"] = df["votes"].str.replace(" ", "", regex=False)

    # Remove extra spaces at start/end
    df["votes"] = df["votes"].str.strip()

    # Convert to numeric
    # If conversion fails, turn value into NaN
    df["votes"] = pd.to_numeric(df["votes"], errors="coerce")


# Step 2: Convert "rank" column to numeric (if it exists)
if "rank" in df.columns:
    df["rank"] = pd.to_numeric(df["rank"], errors="coerce")


# Step 3: Show summary statistics
if "rank" in df.columns:
    print(df[["votes", "rank"]].describe(include="all"))
else:
    print(df[["votes"]].describe())
```

```python
df = df.dropna(subset=["votes"])
df = df[df["votes"] >= 0]
print("After votes cleanup:", df.shape)
```

```python
df = df.drop_duplicates()
df["election_year"] = 2022
print("After dedupe:", df.shape)
```

```python
# Sort by votes descending
df_sorted = df.sort_values(["constituency", "votes"], ascending=[True, False])

# Take top candidate per constituency
winners = df_sorted.groupby("constituency").head(1).copy()

print("Winner rows:", len(winners))
winners.tail(10)
```

```python
df.head(10)
```

```python
# Get top 2 candidates per constituency
top2 = df_sorted.groupby("constituency").head(2)

# Calculate margin
margins = (
    top2.groupby("constituency")["votes"]
    .apply(lambda x: x.iloc[0] - x.iloc[1] if len(x) > 1 else np.nan)
    .reset_index(name="victory_margin")
)

# Merge margin into winners
winners = winners.merge(margins, on="constituency", how="left")

winners.head()
```

```python
winners_path = "../Data/clean/federal_2022_winners.csv"
winners.to_csv(winners_path, index=False, encoding="utf-8-sig")

print("✅ Winners file saved:")
print(winners_path)
```

```python
total_votes = df["votes"].sum()
total_candidates = df["candidate"].nunique()
total_constituencies = df["constituency"].nunique()

print("Total Votes Cast:", total_votes)
print("Total Candidates:", total_candidates)
print("Total Constituencies:", total_constituencies)
```

```python
# Total votes per party + percentage share
party_votes = (
    df.groupby("party")["votes"].sum()
      .reset_index()
)

# Calculate percentage
party_votes["vote_share_%"] = (
    party_votes["votes"] / party_votes["votes"].sum() * 100
)

# Round to 2 decimal places
party_votes["vote_share_%"] = party_votes["vote_share_%"].round(2)

# Sort descending
party_votes = party_votes.sort_values("vote_share_%", ascending=False)

party_votes.head(10)
```

```python
import matplotlib.font_manager as fm

for font in fm.findSystemFonts():
    if "Mangal" in font or "Nirmala" in font:
        print(font)
```

```python
# Bar chart: top 10 parties by vote share

import matplotlib.pyplot as plt
from matplotlib import rcParams

# ✅ Set Nepali-supported font
rcParams["font.family"] = "Nirmala UI"   # or "Mangal"

top10 = party_votes.head(10)

# Make figure wider so bars look bigger
plt.figure(figsize=(10, 6))

# Increase bar width
plt.bar(top10["party"], top10["vote_share_%"], width=0.8)

# Smaller fonts
plt.xticks(rotation=90, fontsize=8)
plt.yticks(fontsize=8)

plt.title("Top 10 Parties by Vote Share (%)", fontsize=10)
plt.ylabel("Vote Share (%)", fontsize=9)
plt.xlabel("Party", fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# Seats won per party (from winners table)
seat_count = (
    winners.groupby("party")
           .size()
           .reset_index(name="seats_won")
)

# Convert to seat share percentage
seat_count["seat_share_%"] = (
    seat_count["seats_won"] / seat_count["seats_won"].sum() * 100
).round(2)

# Join vote share + seat share to compare
comparison = party_votes.merge(seat_count, on="party", how="left").fillna(0)

comparison.sort_values("seats_won", ascending=False).head(10)
```
